```python
import os
import pandas
import datetime
import akshare
import time

current_path = os.path.abspath('')
df = pandas.read_excel(current_path + '/../data/集思录.xlsx')

stock_price = pandas.DataFrame()
stock = pandas.DataFrame()

i = 0
for idx, row in df.iterrows():
    # 第一个是正股代码， 第二个是可转债代码
    symbol = [c.strip() for c in row['证券代码'][1:-1].replace('\'','').split(',')]
    name = [c.strip() for c in row['证券名称'][1:-1].replace('\'','').split(',')]
    # 可转债上市日期
    listing_date = row['上市公告日期']
    # 可转债配售股权登记日, 也就是T-1日（T日是申购日）
    registration = row['股权登记日']
    #print(symbol, name, listing_date, registration)
    registr = datetime.datetime.strptime(registration, "%Y-%m-%d")
    listing_d = datetime.datetime.strptime(listing_date, "%Y-%m-%d")
    
    t_start = (registr - datetime.timedelta(days=30)).strftime("%Y%m%d")
    t_end = (listing_d + datetime.timedelta(days=30)).strftime("%Y%m%d")
    #print(t_start, t_end)
    stock = akshare.stock_zh_a_hist(symbol=symbol[0], start_date=t_start, end_date=t_end)
    stock.insert(0, '股票代码', symbol[0], allow_duplicates=True)
    
    stock_price = pandas.concat([stock_price, stock], ignore_index=True)
    i += 1
    if (i % 50 == 0):
        print("-------------------------------------------------------------------------" + str(i))
        print(stock.head())
        time.sleep(1)
print("-------------------------------------------------------------------------" + str(i))
print(stock.head())

stock_price['均价'] = stock_price.apply(lambda x : x['成交额'] / (x['成交量'] * 100), axis=1)
stock_price.to_excel(current_path + '/../data/正股价格.xlsx', index=False)
print("game over")
```
